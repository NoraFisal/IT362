```python
# Importing cv2 module
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

## Understanding cv2

This is how the actual image looks

```python
#Reading the image
image = mpimg.imread("fruit.jpg")

#Printing the unedited image
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

However, with cv2 the image somehow changes, even though the code is identical?

```python
#Reading the image
image = cv2.imread("fruit.jpg")

#Printing the unedited image
plt.imshow(image)
plt.title("CV2 BGR Channel Image")
plt.axis("off")
plt.show()
```

The reason is that CV2 uses BGR channel and here two of its channels are inverted. We see how to fix this if we are to directly plot off of cv2 below

```python
#Reading the image
image = cv2.imread("fruit.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Printing the unedited image
plt.imshow(image)
plt.title("Original Image with BGR to RBG conversion")
plt.axis("off")
plt.show()
```

As we can see, cv2 has a BRG channel instead of the standard RGB channel!
Now, that we're a bit familiarized with cv2 let's move on to Image processing

## IMAGE SMOOTHENING/ BLURRING

```python
#Reading the image
image = cv2.imread('fruit.jpg')

#This is the gaussian blur we'll be applying at the image
gaussianBlur = cv2.GaussianBlur(image, (15,15),0)

# Plotting and also applying the BGR to RGB transformation!
plt.imshow(cv2.cvtColor(gaussianBlur, cv2.COLOR_BGR2RGB))
plt.title("Image smoothened by Gaussian Blurring")
plt.axis("off")
plt.show()
```

### Averaging

```python
# Averaging
# You can change the kernel size as you want
averaged_image = cv2.blur(image,(10,10))

# Again, you can change the kernel size
plt.imshow(cv2.cvtColor(averaged_image, cv2.COLOR_BGR2RGB))
plt.title("Smoothing by averaging")
plt.axis("off")
plt.show()
```

### Median Blurring

```python
# Median blurring
medBlur = cv2.medianBlur(image,15)
plt.imshow(cv2.cvtColor(medBlur, cv2.COLOR_BGR2RGB))
plt.title("Median Blurring")
plt.axis("off")
plt.show()
```

## THRESHOLDING
We'll see different thresholding techniques and see their outputs

```python
# loading the image
image = cv2.imread('fruit.jpg')

# cv2.cvtColor is applied over the
# image input with applied parameters
# to convert the image in grayscale

image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image, cmap="gray")

plt.title("Gray Scale")
plt.axis("off")
plt.show()
```

Experimenting with various thersholding techniques and graphing their output

```python
# applying different thresholding
# techniques on the input image
# all pixels value above 127 will
# be set to 255. You can try manipulating these values to get different results
ret,thresh1 = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(image,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(image,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(image,127,255,cv2.THRESH_TOZERO_INV)
```

```python
##DISPLAYING THE RESULTANT IMAGES

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [image, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

## EDGE DETECTION

```python
# read the image
image = cv2.imread("fruit.jpg")
```

```python
# convert it to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# show the grayscale image
plt.imshow(gray, cmap="gray")
plt.title("GRAY SCALE")
plt.axis("off")
plt.show()
```

Now we'll detect edges of this gray scale. Compare that image with the gray scale here to see if you notice anything

```python
# perform the canny edge detector to detect image edges
edges = cv2.Canny(gray, threshold1=30, threshold2=100)
# show the resulting image
plt.imshow(edges, cmap="gray")
plt.title("Edges")
plt.axis("off")
plt.show()
```
